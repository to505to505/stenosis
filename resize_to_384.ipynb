```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path

IMAGE_DIR = Path('/home/dsa/stenosis/data/dataset2_split/train/images')
TARGET_SIZE = (384, 384)

# Pick sample images
sample_paths = sorted(IMAGE_DIR.iterdir())[:8]
```

```python
def resize_inter_area(img_path, target_size):
    img = cv2.imread(str(img_path))
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    resized = cv2.resize(img_rgb, target_size, interpolation=cv2.INTER_AREA)
    return img_rgb, resized


fig, axes = plt.subplots(len(sample_paths), 2, figsize=(10, 5 * len(sample_paths)))

for i, path in enumerate(sample_paths):
    original, resized = resize_inter_area(path, TARGET_SIZE)
    axes[i, 0].imshow(original)
    axes[i, 0].set_title(f'Original  {original.shape[1]}x{original.shape[0]}')
    axes[i, 0].axis('off')

    axes[i, 1].imshow(resized)
    axes[i, 1].set_title(f'Resized  {resized.shape[1]}x{resized.shape[0]}  (INTER_AREA)')
    axes[i, 1].axis('off')

    print(f'{path.name}: {original.shape} -> {resized.shape}')

plt.tight_layout()
plt.show()
```
